```python
import os
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
import collections
```

```python
#Importing the NLTK Stopwords
import nltk
nltk.download('stopwords')
stop_list = stopwords.words('english')
print(stop_list)
```

```python
path = 'citeseer'
filenames = os.listdir(path)   #To generate file path
# print(filenames)
```

```python
#Declaring output filenames
collection = 'collection.txt'
tokens_1 = 'tokens_1.txt'
tokens_2 = 'tokens_2.txt'
```

## Task 1

```python
#Parsing all files and merging into a single file collection.txt
with open(collection, 'w') as outfile:
    for fname in filenames:
        filepath = 'citeseer/'+fname
        with open(filepath) as infile:
            outfile.write(infile.read())
        infile.close()
outfile.close()
```

```python
#Tokenizing the text on whitespace and removing punctuation
infile = open(collection,'r')
text = infile.read()
lines = text.lower()                           #converting to lower case
lines = re.sub('[^A-Za-z0-9 ]+',' ',lines)     #removing punctuations by using regular expression
tokens = lines.split()
with open(tokens_1,'w') as outfile:            #writing the tokens into file tokens_1.txt
    for listitem in tokens:
        outfile.write("%s " % listitem)
```

## Task 2

```python
#Determining the frequency of occurrence for all the words in the collection
count = {}
for w in open(tokens_1).read().split():
    if w in count:
        count[w] += 1
    else:
        count[w] = 1
        
print(count)
```

```python
#Calculating total number of words in the collection 
file = open(tokens_1)
data = file.read()
words = data.split()
len(words)
```

```python
#Calculating vocabulary size (number of unique words)
len(count)
```

```python
#Sorting in decreasing order to get most frequent words first 
count = {k: v for k, v in sorted(count.items(), key=lambda item: item[1], reverse = True)}

#Finding the top 20 words in the collection
top20 = {k: v for k, v in sorted(count.items(), key=lambda item: item[1], reverse = True)[:20]}
print(top20)
print(top20.keys())
```

```python
#Finding the stop words in the top 20 words
k = top20.keys()
for i in k:
    if i in stop_list:
        print(i,' ')
```

```python
#Determining minimum number of unique words accounting for 15% of the total number of words in the collection
c = 0
n =0
limit = len(words)*0.15
for i in count:
    if c < limit and count[i] <= (limit - c):
        print(i,':', count[i])
        c += count[i]
        n += 1   
print('\nNumber of words:',n)
print('Total count:',c)
```
